# Network Infodemic Model
---

This notebook is an example of how to use the model.

Start by importing the model's modules, and matplotlib for plotting support later on.

```python
import matplotlib.pyplot as plt

import nim.core
import nim.plot
```

With the modules imported, you can access all of the model's functionality. To run a model, invoke `NetworkModel` from the `nim.core` module to create an instance of the model object. The object will store all the data and serve as the model interface.

```python
nimodel = nim.core.NetworkModel(
    number_of_nodes=100,
    interaction_type="symmetric")
```

With the model object instantiated, you can use its attached methods to perform operations. For example, if you wanted to simulate with a GNP-random network as your basis, you could do that like so:

```python
nimodel.create_network(network_type="gnp_random", p=0.3)
```

With the basis network initialized on the model object, you can now run a simulation.

```python
nimodel.initialize_connections()
nimodel.initialize_attitudes()
nimodel.run_simulation(tmax=100)
```

Once the simulation is complete, the model object can be passed to `NetworkPlot` from the `nim.plot` module to see how the simulation went. This module allows quick access to simple aspects of the simulation results.

```python
niplot = nim.plot.NetworkPlot(nimodel)
```

You can examine how the mean node attitudes and mean connection strength evolved over time:

```python
means_over_time = niplot.attitude_and_connection_means()
plt.show(means_over_time)
```

And you can view how each individual node's attitude evolved over time:

```python
attitudes_over_time = niplot.node_attitudes_over_time()
plt.show(attitudes_over_time)
```

For large networks, e.g., 1000+ nodes, this individual node plot will be slow and unhelpful, but it can be used with a smaller network to give an illustrative example to pair with the plot of the means earlier.
